```python
def journal_figure(do_save=False, filename='figure.eps', dpi=300, size_inches=(2.16, 2.16), linewidth=1):
    """
    Adjusts the current matplotlib figure to make it look publication-worthy.
    
    Parameters:
    - do_save: bool, whether to save the figure to an EPS file.
    - filename: str, the name of the file to save the figure as.
    - dpi: int, the resolution of the figure in dots per inch.
    - size_inches: tuple, the size of the figure in inches.
    - linewidth: float, the line width for the plot elements.
    """
    ax = plt.gca()  # Get the current axes
    
    # Adjust tick direction and length
    ax.tick_params(direction='out', length=10, width=linewidth)
    
    # Turn off the top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(direction='out', length=6, width=2)
    ax.set_aspect(1.0/plt.gca().get_data_ratio(), adjustable='box')
    # Set font size and type
    plt.xticks(fontsize=12, fontname='Times New Roman')
    plt.yticks(fontsize=12, fontname='Times New Roman')
    
    if do_save:
        # Save the figure
        plt.savefig(filename, dpi=dpi, bbox_inches='tight', format='eps', linewidth=linewidth)
```

```python
def print_ranksum(group1, group2):
    from scipy.stats import ranksums
    stat, p = ranksums(group1, group2)
    df = len(group1) + len(group2) - 2
    if p < 0.001:
        print(f"U = {stat:.3f}, p < 0.001")
    elif p < 0.05:
        print(f"U = {stat:.3f}, p = {p:.3f}")
    else:   
        print(f"U = {stat:.3f}, p = {p:.3f}")
    return stat, p
```

```python
def new_journal_figure(ax=None, do_save=False, filename='figure.eps', dpi=300, size_inches=(2.16, 2.16), linewidth=1, set_aspect=False):
    """
    Adjusts a matplotlib axis to make it look publication-worthy.
    """
    if ax is None:
        ax = plt.gca()

    ax.tick_params(direction='out', length=6, width=linewidth)

    # Turn off the top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Only set aspect if explicitly asked (not EPS-safe with multiple axes)
    if set_aspect:
        try:
            ax.set_aspect(1.0 / ax.get_data_ratio(), adjustable='box')
        except Exception as e:
            print("Aspect ratio setting skipped due to:", e)

    # Set font style
    for label in (ax.get_xticklabels() + ax.get_yticklabels()):
        label.set_fontsize(12)
        label.set_fontname('Times New Roman')

    if do_save:
        plt.savefig(filename, dpi=dpi, bbox_inches='tight', format='eps', linewidth=linewidth)
```

```python
def print_ttest_ind(group1, group2):
    stat, p = ttest_ind(group1, group2, equal_var=False)
    df = len(group1) + len(group2) - 2
    print(f"t({df}) = {stat:.3f}, p = {p:.3f}")
    return stat, p, df
```

```python
from scipy.stats import spearmanr, pearsonr
def print_correlation(x, y):
    if np.isnan(x).any():
        x = np.nan_to_num(x)
    if np.isnan(y).any():
        y = np.nan_to_num(y)
    r, p = spearmanr(x, y)
    d = len(x)-2
    print(f"r({d}) = {r:.3f}, p = {p:.3f}")
    return r, p
```

```python
import numpy as np
from sklearn.decomposition import PCA

def effective_dim_by_variance(features, variance_threshold=0.9):
    """
    Computes the number of principal components needed to explain variance_threshold.
    
    Parameters:
        features: np.ndarray, shape (n_samples, n_features)
        variance_threshold: float, e.g. 0.9 for 90% explained variance

    Returns:
        k: int, number of components
    """
    X = features - np.mean(features, axis=0)
    pca = PCA()
    pca.fit(X)
    cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
    k = np.searchsorted(cumulative_variance, variance_threshold) + 1
    return k
```

```python
def participation_ratio(features):
    X = features - np.mean(features, axis=0)
    pca = PCA()
    pca.fit(X)
    eigvals = pca.explained_variance_
    return (np.sum(eigvals) ** 2) / np.sum(eigvals ** 2)
```

```python
def print_wilcoxon(group1, group2):
    stat, p = scipy.stats.wilcoxon(group1, group2, alternative='greater')
    d = len(group1)
    print(f"Z = {stat:.3f}, p = {p:.3f}")
    return stat, p, d
```

```python
def print_ttest_rel(group1, group2):
    stat, p = scipy.stats.ttest_rel(group1, group2)
    d = len(group1) - 1
    print(f"t({d}) = {stat:.3f}, p = {p:.3f}")
    return stat, p, d
```

```python
from scipy.stats import wilcoxon, ranksums
import numpy as np
import pandas as pd

ssl_models_list = ["resnet", "convnext", "vit", "resnet101", "swin"]
monkeys = ["magneto", "nano"]

results = []

for monkey in monkeys:
    for model in ssl_models_list:
        if model == "resnet":
            model_name = "resnet"
            ssl_model_name = "resnet_swsl"
        else:
            model_name = model
            ssl_model_name = f"{model}_ssl"

        ev_model = np.load(f"./results_predictions/monkey2model/ssl_models_v2/{monkey}/ev_{model_name}.npy")
        ev_ssl = np.load(f"./results_predictions/monkey2model/ssl_models_v2/{monkey}/ev_{ssl_model_name}.npy")
        
        # Remove NaNs and ensure same size
        if ev_model.shape[0] != ev_ssl.shape[0]:
            ev_model = ev_model[:ev_ssl.shape[0]]
        mask = ~np.isnan(ev_model) & ~np.isnan(ev_ssl)
        ev_model_clean = ev_model[mask]
        ev_ssl_clean = ev_ssl[mask]

        try:
            if len(ev_model_clean) > 0 and not np.allclose(ev_ssl_clean, ev_model_clean):
                stat, p = ranksums(ev_ssl_clean, ev_model_clean, alternative="greater")
            else:
                stat, p = np.nan, np.nan
        except ValueError as e:
            stat, p = np.nan, np.nan

        results.append({
            "Monkey": monkey,
            "Model": model,
            "Mean SL": np.nanmean(ev_model),
            "Mean SSL": np.nanmean(ev_ssl),
            "N Samples": len(ev_model_clean),
            "Wilcoxon p-value (SSL > SL)": p
        })

# Save or print results
df = pd.DataFrame(results)
print(df)
df.to_csv("ssl_vs_sl_stats.csv", index=False)
```
